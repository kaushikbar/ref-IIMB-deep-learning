```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
num_inputs = 4
```

```python
# XOR
input_array = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1],
])
output_array = np.array([0, 1, 1, 0])
```

```python
ones = np.ones((num_inputs, 1))
input_bias = np.hstack((ones,  input_array))
```

### ANN


```python
# Importing the Keras libraries and packages
import keras  
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(output_dim = 4, activation = 'sigmoid', input_dim = 3))

# Adding the output layer
classifier.add(Dense(output_dim = 1, activation = 'sigmoid'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier.fit(input_bias, output_array, nb_epoch = 3000)
```

```python
# Part 3 - Making predictions and evaluating the model

y_pred = classifier.predict(input_bias)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_pred, output_array)
cm
```
